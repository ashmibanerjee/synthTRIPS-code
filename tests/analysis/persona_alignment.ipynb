```python
import pandas as pd 
import numpy as np 
import os 
import re 
import json 
import sys
```

```python
llama = pd.read_csv("../../data/conv-trs/eval/persona-alignment/llama_alignment_gpt.csv")
```

```python
llama.head()
```

```python
mapping = {
    r"^(A: )?Not Aligned.*|^A$": "A: Not Aligned",
    r"^(B: )?Partially Aligned.*|^B$": "B: Partially Aligned",
    r"^(C: )?Aligned.*|^C$": "C: Aligned",
    r"^(D: )?Unclear.*|^D$": "D: Unclear"
}
```

```python
# def clean_result(df):
for method in ['v', 'p0', 'p1']:
    llama[f'explanation_{method}'] = llama[f'query_{method}_matches'].str.extract(r'Evaluation:\s*(.*?)\s*Rating:', flags=re.DOTALL)
    llama[f'rating_{method}'] = llama[f'query_{method}_matches'].str.extract(r'Rating:\s*(.*)')
    # Apply regex-based replacement
    llama[f"rating_{method}"] = llama[f"rating_{method}"].replace(mapping, regex=True)
```
